# **MY FIRST ML PROJECT**

# **LOAD DATA**

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

# **DATA PREPRATION**

## DATA SEPARATION AS X & Y

```python
y = df['logS']
y
```

```python
x = df.drop('logS' , axis=1)
x
```

## data spliting

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=98)
```

```python
x_train
```

```python
x_test
```

# **MODEL BULIDING**

## Linear regression

### **training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

### **applying the model to make pedictions**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

### evaluate model performance

```python
from sklearn.metrics import mean_squared_error , r2_score
lr_train_mse = mean_squared_error(y_train , y_lr_train_pred)
lr_train_r2 = r2_score(y_train , y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test , y_lr_test_pred)
lr_test_r2 = r2_score(y_test , y_lr_test_pred)
```

```python
print('LR MSE (Train): ' , lr_train_mse)
print('LR R2 (Train): ' , lr_train_r2)
print('LR MSE (Test): ' , lr_test_mse)
print('LR R2 (Test): ' , lr_test_r2)
```

```python
lr_results = pd.DataFrame(['linear regression' , lr_train_mse , lr_train_r2 , lr_test_mse , lr_test_r2]).transpose()
lr_results.columns = ['algorithm' , 'trainingc mse' , 'training r2' , 'testing mse' , 'testing r2']
```

```python
lr_results
```

## **RANDOM FOREST**

### TRAINING THE MODEL

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2 , random_state=98)
rf.fit(x_train , y_train)
```

### APPLYING THE MODEL TO MAKE PREDICTION

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### EVALUATING THE MODEL PERFORMANCE

```python
from sklearn.metrics import mean_squared_error , r2_score
rf_train_mse = mean_squared_error(y_train , y_rf_train_pred)
rf_train_r2 = r2_score(y_train , y_rf_train_pred)
rf_test_mse = mean_squared_error(y_test , y_rf_test_pred)
rf_test_r2 = r2_score(y_test , y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['algorithm', 'trainingc mse', 'training r2', 'testing mse', 'Training MSE']
rf_results
```

# **MODEL COMAPRISON**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```

# **DATA VISUALIZATION OF PREDICION RESULTS**

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,5))
plt.scatter(x=y_train , y=y_lr_train_pred , c='g' , alpha=0.3)
z = np.polyfit(y_train , y_lr_train_pred , 1)
p = np.poly1d(z)
plt.plot(y_train , p(y_train) , c='r')
plt.ylabel('predicted logS')
plt.xlabel('Experimental logS')
```
